```python
# fill numerical values with mean
means = df.select_dtypes(include=np.number).mean()
df.fillna(means, inplace=True)
# fill categorical values with mode
mode = df.select_dtypes(exclude=np.number).mode()
modes = mode.iloc[0]
df[mode.columns] = df[mode.columns].fillna(modes)

# 3. Check if there are any null values left
df.isnull().sum()
```

# **Overall disease ratio**

```python
import matplotlib.pyplot as plt
colors = ['#ff6666', '#3399ff', '#99ff99', '#ffcc66', '#9966ff']
plt.figure(figsize=(7, 7))
mylabels = ['Hepatitis B', 'HIV/AIDS', 'Measles ', 'Polio', 'Diphtheria ']
myexplode = [0.07, 0.1, 0.07, 0.07, 0.07]

values = [df['Hepatitis B'].sum(), df[' HIV/AIDS'].sum(), df['Measles '].sum(),
          df['Polio'].sum(), df['Diphtheria '].sum()]
wedges, texts, autotexts = plt.pie(values,
                                   labels=mylabels,
                                   explode=myexplode,
                                   shadow=True,
                                   autopct="%1.1f%%",
                                   startangle=140,
                                   colors=colors,
                                   wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'alpha': 0.8})
for text in texts:
    text.set_fontsize(12)
    text.set_color('black')
for autotext in autotexts:
    autotext.set_fontsize(11)
    autotext.set_color('white')
    autotext.set_weight('bold')

plt.title("Overall Disease Ratio", fontsize=18, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# **Highest Alcohol consumption**

```python
top_20_alcohol_countries = df.groupby('Country')['Alcohol'].sum().nlargest(20)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_20_alcohol_countries.index, y=top_20_alcohol_countries.values, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Country')
plt.ylabel('Total Alcohol Consumption')
plt.title('Top 20 Countries with Highest Alcohol Consumption', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

## **Life expectancy of top 20 countries in our dataset**

```python
country_counts = df["Country"].value_counts().head(20)

top_countries_df = df[df["Country"].isin(country_counts.index)]

plt.figure(figsize=(12, 8))
sns.countplot(x='Country', data=top_countries_df, order=country_counts.index, palette='viridis')

plt.xticks(rotation=45, ha="right")
plt.title("Top 20 Countries by Life Expectancy")
plt.tight_layout()
plt.show()
```

# **Adult mortality**

```python
top_20_adult_mortality = df.groupby('Country')['Adult Mortality'].mean().nlargest(20)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_20_adult_mortality.index, y=top_20_adult_mortality.values, palette='coolwarm')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Country')
plt.ylabel('Adult Mortality Ratio')
plt.title('Top 20 Countries with Highest Adult Mortality Ratio', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

# **Economic indicators of a random country in our dataset**

```python
country_name = 'Brazil'
country_data = df[df['Country'] == country_name]
total_expenditure = country_data['Total expenditure'].sum()
# Check for typos and correct the column name or use an alternative method to calculate percentage expenditure.
percentage_expenditure = country_data['percentage expenditure'].mean()
gdp = country_data['GDP'].mean()
labels = ['Total Expenditure', 'Percentage Expenditure', 'GDP']
sizes = [total_expenditure, percentage_expenditure, gdp]
colors = ['#ff9999', '#66b3ff', '#99ff99']
explode = (0.1, 0, 0)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title(f'Economic Indicators for {country_name}', fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```
